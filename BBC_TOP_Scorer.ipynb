```python
Player_Names = []
Team_name = []
Goals = []
Assists = []
No_of_matches = []
Shots = []  
```

```python
players = soup.find("tbody").find_all("tr", class_="ssrcss-dhlz6k-TableRowBody e1icz100")
for player in players:
    player_name =player.find("div", class_="ssrcss-m6ah29-PlayerName e1n8xy5b1").get_text(strip=True)
    team = player.find("div", class_ ="ssrcss-qvpga1-TeamsSummary e1n8xy5b0").get_text(strip=True)
    goal_scored=player.find("div", class_="ssrcss-8k20kk-CellWrapper ef9ipf0").get_text(strip=True)
    stats = player.find_all("div",class_="ssrcss-150z8d-CellWrapper ef9ipf0") # 10 cells in this class
    # for div_tag in stats:
    #     print(div_tag)
    assists_made=int(stats[0].get_text(strip=True))
    matches_played =int( stats[2].get_text(strip=True))
    shots_taken=int(stats[-3].get_text(strip=True))
    
    Player_Names.append(player_name)
    Team_name.append(team)
    Goals.append(goal_scored)
    Assists.append(assists_made)
    No_of_matches.append(matches_played)
    Shots.append(shots_taken)

```

```python
data={
    "Player":Player_Names,
    "Team":Team_name,
    "Goals":Goals,
    "Assists":Assists,
    "Matches":No_of_matches,
    "Shots":Shots
    
}
```

```python
df_players=pd.DataFrame(data)
print(df_players)
```

```python
df_players.to_excel("Premier league Top Scorers.xlsx",index=False)
```
